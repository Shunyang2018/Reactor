```python
from rdkit import Chem
from rdkit.Chem import rdChemReactions
```

## glucose

### reference: https://www.daylight.com/dayhtml_tutorials/languages/smarts/
1) \#7 all N  
2) #7H1 N with 1 Hydrogen  
3) c aromatic C  
4) C aliphatic C  
5) , : or 

```python
rx = '[#8H1,#7H1,#7H2,c:1].OCC1O[C:5](O)C(O)C(O)C1O>>OCC1O[CH:5]([#7H1,#7H2,#8,c:1])C(O)C(O)C1O' 
#'[*:1].OCC1O[CH:5]C(O)C(O)C1O>>OCC1O[CH:5]([*:1])C(O)C(O)C1O'
```

```python
rxn = rdChemReactions.ReactionFromSmarts(rx)
```

```python
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.molSize = 250,250
```

```python
glu = '[H]O[CH:5]1OC(CO)C(O)C(O)C1O'
r = '[H][CH]=CCC(N)C(O)=O'
reacts = (Chem.MolFromSmiles(r),Chem.MolFromSmiles(glu))
products = rxn.RunReactants(reacts)
```

```python
rxn.RunReactants(reacts)
```

```python
def glyco(r,glu,rx):
    '''
    r: reactant smiles
    glu: sugar smiles
    rx: reaction smarts
    '''
    rxn = rdChemReactions.ReactionFromSmarts(rx)
    reacts = (Chem.MolFromSmiles(r),Chem.MolFromSmiles(glu))
    products = rxn.RunReactants(reacts)
    return Chem.MolToSmiles(products[0][0]) # return only one products, O>N>C
```

```python
Chem.MolToSmiles(glyco(r,glu,rx))
```
